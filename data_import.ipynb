# Cryptocurrency Analysis  

## 1. Let's first import the coins' data from coincap's API, and save the data as a dataframe.

```python
# Importing pandas
import pandas as pd
import requests

# Reading in current data from coinmarketcap.com
r=requests.get("http://coincap.io/front")
# or dec6 = pd.read_csv("https://raw.githubusercontent.com/superbunny/coinmarket/master/coinmarketcap_06122017.csv")
j=r.json()
data=pd.DataFrame(j)

# filter the data to get observations with higher-than-zero market cap. 
cap = data.query('mktcap > 0')

#Take a look at the data  
cap
```

```python
cap.count() #Take a look at the column names and the number of observations
```

## 2. How big is Bitcoin compared with the rest of the cryptocurrencies?  

```python
#Plot the market capitalization for the top 10 coins.
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg' 
plt.style.use('fivethirtyeight')

cap=cap.rename(columns={'long':"ID"})
cap10 = cap[:10].set_index('ID')
cap10=cap10.assign(market_cap_perc=lambda x: (x.mktcap/cap.mktcap.sum())*100)
ax=cap10['market_cap_perc'].plot.bar()
ax.set_title('Top 10 market capitalization')
ax.set_ylabel('% of total cap')
ax.set_xlabel('')
```
